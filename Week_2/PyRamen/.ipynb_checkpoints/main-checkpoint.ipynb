```python
import pandas as pd

import numpy as np
```

```python
# Read the data and set it into a variable

menu_df = pd.read_csv('menu_data.csv')

sales_df = pd.read_csv('sales_data.csv', index_col = "Line_Item_ID", parse_dates = True, infer_datetime_format = True)

sales_df.head(10)
```

```python
sales_df.duplicated().sum()
```

```python
sales_df.isnull().sum()
```

```python
sales_df.drop_duplicates(inplace=True)
```

```python
sales_df.duplicated().sum()
```

```python
sales_df.head(10)
```

```python
sales_df['Menu_Item'].value_counts()
```

```python
sales_sliced_df = sales_df.groupby(['Menu_Item']).count()
```

```python
menu_sliced_df = menu_df.loc[14:24, :]
```

```python
sales_sliced_df.drop(columns = ['Date', 'Credit_Card_Number'], inplace=True)
```

```python
sales_sliced_df.sort_values("Menu_Item", ascending = True, inplace = True)

sales_sliced_df
```

```python
menu_sliced_df.sort_values("item", ascending = True, inplace = True)

menu_sliced_df
```

```python
menu_sliced_df.reset_index(inplace=True)
```

```python
menu_sliced_df.drop(columns = 'index', inplace = True)
```

```python
menu_sliced_df
```

```python
sales_sliced_df.reset_index(inplace=True)
```

```python
sales_sliced_df
```

```python
total_sales_df = pd.concat([menu_sliced_df, sales_sliced_df], axis = 'columns', join = 'inner')

total_sales_df
```

```python
total_sales_df.drop(columns = 'Menu_Item', inplace = True)

total_sales_df
```

```python
total_sales_df.columns = ['Menu_Item', 'Category', 'Description', 'Price', 'Cost','Count']

total_sales_df
```

```python
total_sales_df['Revenue'] = total_sales_df['Price']*total_sales_df['Count']

total_sales_df
```

```python
total_sales_df['COGS'] = total_sales_df['Cost'] * total_sales_df['Count']
```

```python
total_sales_df
```

```python
total_sales_df['Profit'] = total_sales_df['Revenue'] - total_sales_df['COGS']

total_sales_df
```

```python
total_sales_df.to_csv('total_sales_data.csv')
```
